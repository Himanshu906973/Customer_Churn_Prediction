## 14. Feature Encoding (Categorical to Numerical)

```python
df = pd.get_dummies(df, columns=['Geography','Gender'], drop_first=True, dtype=int)
```

## 15. Processed Dataset

```python
df
```

## 16. Splitting Dataset into Training and Testing Sets

```python
X=df.drop(columns=['Exited'])
y=df['Exited']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2, random_state=1)
```

## 17. Training Data Size

```python
X_train.shape
```

## 18. Feature Scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## 19. Import TensorFlow and Keras

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

## 20. Building Artificial Neural Network Model

```python
model = Sequential()

model.add(Dense(3,activation='sigmoid',input_dim=11))
model.add(Dense(1,activation='sigmoid'))
```

## 21. Model Architecture Summary

```python
model.summary()
```

## 22. Compile the Model

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

## 23. Model Training

```python
model.fit(X_train,y_train,epochs=10)
```

## 24. Extract Model Weights

```python
model.layers[0].get_weights()
```

```python
model.layers[1].get_weights()
```

## 25. Predicting Customer Churn

```python
y_log=model.predict(X_test_scaled)
```

## 26. Converting Probabilities to Binary Predictions

```python
import numpy as np
y_pred=np.where(y_log>0.5, 1, 0)
```

## 27. Model Accuracy Evaluation

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# 28. Confusion Matrix
